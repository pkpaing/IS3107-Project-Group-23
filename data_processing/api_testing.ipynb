```python
import pandas as pd
import requests
import json
```

```python
url = 'http://103.195.4.105:8000/bus_stop'

resp = requests.get(url)

data = pd.DataFrame(resp.json())

data['BusStopCode'] = data['BusStopCode'].apply(lambda x: x * 1000)

data
```

```python
data_to_write = json.loads(data.to_json(orient='records'))

local_url = 'http://127.0.0.1:8000/bus_stops/'

# Make the POST request
response = requests.post(local_url, json=data_to_write)

print(response.json())
```
